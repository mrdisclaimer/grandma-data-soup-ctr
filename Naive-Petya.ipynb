```python
import pandas as pd
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
test.shape, test.drop('RowId', axis=1).drop_duplicates().shape
```

```python
cols = [u'CampaignId', u'AdGroupId', u'KeywordId', u'Device', u'AveragePosition']
```

```python
ctr = train.groupby(cols).sum()
ctr['ClickProbability'] = ctr.Clicks / ctr.Impressions

sub = test.set_index(cols).join(ctr, how='left').reset_index()

# Заполняем средним
sub['ClickProbability'].fillna(train.Clicks.sum()/train.Impressions.sum(), inplace=True)
```
